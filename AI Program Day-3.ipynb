```python
#Find total sales amount by product category.
import pandas as pd 
import numpy as mp
import matplotlib.pyplot as plt
df=pd.read_csv("Amazon Sales Data.csv",encoding='unicode_escape')
sales_prod=df.groupby(["Product_Category"],as_index=False)["Amount"].sum()
sales_prod
```

```python
import seaborn as sns
sales_prod=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(ascending=True,by="Amount").head()
sales_prod
sns.barplot(x="Product_Category",y="Amount",data=sales_prod)
```

```python
#Write a function that will give count of vowels present in the string/text.
#Not to use any predefined python libraries/function.
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
text = "This is TMV college ,formerly University college For Women,Koti"
print("Number of vowels:", count_vowels(text))
```

```python
#Write a function that will give count of vowels present in the string/text.
#Not to use any predefined python libraries/function.
def count_vowels(x):
    c = 0 
    v = ['a','e','i','o','u']    
    for i in x:
        if i in v:
            c += 1
    return c
count_vowels("Telangana Mahila Vishwavidhyalayam")
```
